```python
#MÃ‰DIA
exemplo_series = pd.Series([1,5,10,30,50,30,15,40,45])
print(titanic['Age'].mean())
print(exemplo_series.mean())
```

```python
#MEDIANA
print(titanic['Age'].median())
print(exemplo_series.median())
```

```python
# QUANTIL
print(titanic['Age'].quantile())
print(exemplo_series.quantile())
```

```python
# QUANTIL
print(titanic['Age'].quantile(q=0.25))
print(exemplo_series.quantile(q=0.25))
```

```python
#MODA
print(titanic['Age'].mode())
print(exemplo_series.mode())
```

```python
#DOIS VALORES DIFERENTES PARA MODA
exemplo_series = pd.Series([1,5,10,30,50,30,15,40,45,45])
print(exemplo_series.mode())
```

```python
#MEDIDAS DE DISPERSÃƒO - AMPLITUDE
print(titanic['Age'].max() - titanic['Age'].min())
print(exemplo_series.max() - exemplo_series.min())
```

```python
#VARIÃ‚NCIA
print(titanic['Age'].var())
print(exemplo_series.var())
```

```python
#DESVIOPADRÃƒO
print(titanic['Age'].std())
print(exemplo_series.std())
```

```python
#DESVIABSOLUTO
print(titanic['Age'].mad())
print(exemplo_series.mad())
```
